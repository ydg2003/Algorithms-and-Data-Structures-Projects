1.1) Dada las siguientes definiciones formales de la estructura de grafos, provea una representación grafica
de los mismos. (15 puntos)

a) V(G) = {1,2,3,4,5,6,7,8}
E(G) = { {1,2},{2,3},{2,6},{3,6},{4,7},{6,8},{7,8} }

b) V(G) = {1,2,3,4,5,6,7}
E(G) = { {1,2},{2,3},{3,4},{3,5},{4,5},{5,6},{6,7} }

c) V(G) = {1,2,3,4,5,6,7,8,9}
E(G) = { {1,2},{1,4},{2,3},{2,5},{2,6},{3,6},{3,9},{4,7}, {5,6},{5,7},{6,8},{6,9},{7,8},{8,9} }

```python
import networkx as nx
import matplotlib.pyplot as plt

# Graph a)
vertices_a = {1, 2, 3, 4, 5, 6, 7, 8}
edges_a = [(1, 2), (2, 3), (2, 6), (3, 6), (4, 7), (6, 8), (7, 8)]
G_a = nx.Graph()
G_a.add_nodes_from(vertices_a)
G_a.add_edges_from(edges_a)

# Graph b)
vertices_b = {1, 2, 3, 4, 5, 6, 7}
edges_b = [(1, 2), (2, 3), (3, 4), (3, 5), (4, 5), (5, 6), (6, 7)]
G_b = nx.Graph()
G_b.add_nodes_from(vertices_b)
G_b.add_edges_from(edges_b)

# Graph c)
vertices_c = {1, 2, 3, 4, 5, 6, 7, 8, 9}
edges_c = [(1, 2), (1, 4), (2, 3), (2, 5), (2, 6), (3, 6), (3, 9), (4, 7), (5, 6), (5, 7), (6, 8), (6, 9), (7, 8), (8, 9)]
G_c = nx.Graph()
G_c.add_nodes_from(vertices_c)
G_c.add_edges_from(edges_c)

# Draw graphs
plt.figure(figsize=(18, 4))

plt.subplot(131)
nx.draw(G_a, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=8)
plt.title('Graph a)')

plt.subplot(132)
nx.draw(G_b, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=8)
plt.title('Graph b)')

plt.subplot(133)
nx.draw(G_c, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=8)
plt.title('Graph c)')

plt.show()
```

1.3) Dado el siguiente Grafo, realice las operaciones solicitadas. (25 puntos)

a) Provea una definición formal de la estructura del Grafo en el formato G=(V,E)

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define vertices and edges for Graph z)
vertices_z = {'v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'v7', 'v8'}
edges_z = [('v1', 'v2'), ('v1', 'v5'), ('v2', 'v8'), ('v5', 'v3'), ('v3', 'v4'), ('v4', 'v5'),
           ('v6', 'v5'), ('v2', 'v5'), ('v1', 'v6'), ('v6', 'v7'), ('v5', 'v7'), ('v6', 'v8'),
           ('v7', 'v4'), ('v8', 'v7'), ('v8', 'v4'), ('v2', 'v3'), ('v1', 'v7'), ('v5', 'v8')]

# Create a graph for Graph z)
G_z = nx.Graph()

# Add vertices to the graph
G_z.add_nodes_from(vertices_z)

# Add edges to the graph
G_z.add_edges_from(edges_z)

# Draw the graph
nx.draw(G_z, with_labels=True, font_weight='bold', node_size=700, node_color='lightgreen', font_size=8)

# Show the graph
plt.show()
```

b) Calcule el orden y el tamaño del grafo

```python
# Order of the graph (number of vertices)
order = len(G_z.nodes())

# Size of the graph (number of edges)
size = len(G_z.edges())

print("Order of the graph:", order)
print("Size of the graph:", size)
```

c) Para cada vértice (nodo) liste sus vecinos N(v) y calcule su grado d(v)

```python
# Calculate and print neighbors and degree for each vertex
for vertex in G_z.nodes():
    neighbors = list(G_z.neighbors(vertex))
    degree = G_z.degree(vertex)
    print(f"Vertex {vertex}: Neighbors {neighbors}, Degree {degree}")
```

d) Calcule la matriz de adyacencia y de incidencia del grafo

```python
# Calculate adjacency matrix
adjacency_matrix = nx.adjacency_matrix(G_z).todense()
print("\nAdjacency Matrix:")
print(adjacency_matrix)

# Calculate incidence matrix
incidence_matrix = nx.incidence_matrix(G_z).todense()
print("\nIncidence Matrix:")
print(incidence_matrix)
```
