## Part2 a) QuickSort Algorithm for random pivoting using Hoare Partitioning

```python
# Python implementation QuickSort using 
# Hoare's partition Scheme.

import random
def quicksort(A, l, r):
	if(l < r):
		
		# pivotindex is the index where
		# the pivot lies in the A
		pivotindex = partitionrand(A,\
							l, r)
		
		# At this stage the A is 
		# partially Sorted Arrayround the pivot. 
		# separately sorting the left half of 
		# the A and the right
		# half of the A.
		quicksort(A , l , pivotindex)
		quicksort(A, pivotindex + 1, r)

# This function generates random pivot,
# swaps the first element with the pivot
# and calls the partition function.
def partitionrand(A , l, r):

	# Generating a random number between 
	# the ling index of the A and 
	# the ending index of the A.
	randpivot = random.randrange(l, r)

	# Swapping the ling element of 
	# the A and the pivot
	A[l], A[randpivot] =\
		A[randpivot], A[l]
	return partition(A, l, r)

def partition(A,l,r):
	pivot = l # pivot
	i = l - 1
	j = r + 1
	while True:
		while True:
			i = i + 1
			if A[i] >= A[pivot]:
				break
		while True:
			j = j - 1
			if A[j] <= A[pivot]:
				break
		if i >= j:
			return j
		A[i] , A[j] = A[j] , A[i]

# Driver Code
if __name__ == "__main__":
	A = [10, 7, 8, 9, 1, 5]
	quicksort(A, 0, len(A) - 1)
	print(A)
```
